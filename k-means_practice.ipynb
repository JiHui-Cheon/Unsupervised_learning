```python
# 가우스함수 생성
def gauss(x, mu, sigma):
    N, D = x.shape
    c1 = 1 / (2*np.pi)**(D / 2)
    c2 = 1 / (np.linalg.det(sigma)**(1 / 2)) # 행렬식
    inv_sigma = np.linalg.inv(sigma) # 역행렬
    c3 = x - mu
    c4 = np.dot(c3, inv_sigma)
    c5 = np.zeros(N)
    for d in range(D):
        c5 = c5 + c4[:, d] * c3[:, d]
#     p = c1 * c2 * np.exp(-c5 / 2)
    return c5
```
